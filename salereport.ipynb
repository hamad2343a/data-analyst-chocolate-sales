```python
import pandas as pd

# Replace the link below with your raw CSV link from GitHub
url = "https://raw.githubusercontent.com/hamad2343a/data-analyst-chocolate-sales/refs/heads/main/customers.csv"

data = pd.read_csv(url)
print(data.head())

```

```python
# Check basic info
print(data.info())

# Check first 5 rows
print(data.head())

# Summary statistics for numerical columns
print(data.describe())
```

```python
print(data.isnull().sum())
# Drop missing rows
data = data.dropna()
```

```python
data['join_date'] = pd.to_datetime(data['join_date'])
total_customers = data.shape[0]
print("Total customers:", total_customers)
```

```python
#Gender Distribution
gender_counts = data['gender'].value_counts()
print(gender_counts)
```

```python
#Average age of customer
avg_age = data['age'].mean()
print("Average customer age:", avg_age)
```

```python
loyalty_counts = data['loyalty_member'].value_counts()
print(loyalty_counts)
data['join_date'] = pd.to_datetime(data['join_date'])
```

```python
import matplotlib.pyplot as plt

# Gender distribution
gender_counts.plot(kind='bar', color=['blue','pink'])
plt.title('Customer Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()

# Loyalty members
loyalty_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Loyalty Membership')
plt.show()

# Age distribution
data['age'].plot(kind='hist', bins=5, color='green')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```
